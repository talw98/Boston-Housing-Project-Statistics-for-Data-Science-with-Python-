## TASK#6 Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)

First we will create a categorical column which will tell us whether the house is FAR or NEAR by converting '0 and 1' in the existing 'CHAS' column of the dataset

```python
boston_df.loc[(boston_df['CHAS'] == 0), 'CHAS_T'] = 'FAR'
boston_df.loc[(boston_df['CHAS'] == 1), 'CHAS_T'] = 'NEAR'

#now let's ptint the modified dataset
boston_df.head(5)
```

Null Hypothesis = 'There is no significant difference in median value of houses bounded by Charles River'


Alternative Hypothesis = 'There is a significant difference in median value of houses bounded by Charles River'

```python
scipy.stats.ttest_ind(boston_df[boston_df['CHAS_T'] == 'FAR']['MEDV'],
                          boston_df[boston_df['CHAS_T'] == 'NEAR']['MEDV'], equal_var=True)
```

Since p-value is less than 0.05(as we set our significance level at 0.05), we can reject null hypothesis and conclude *'there is indeed a significant
difference in median value of houses bounded by Charles River.*'

## TASK#7 Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)? (ANOVA)



Null Hypothesis = There is no difference in Median values of houses for proportion owner occupied units built prior to 1940 (AGE)

Alternate Hypothesis = There is a significant difference in Median values of houses for proportion owner occupied units built prior to 1940 (AGE)


```python
scipy.stats.f_oneway(boston_df['MEDV'], (boston_df['AGE'] < 1940))
```

Since p-value is less than 0.05, we can reject the null hypothesis and claim *'there is a significant difference in Median values of houses for proportion owner occupied units built prior to 1940 (AGE).*'

## TASK#8 Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)



Null Hypothesis = There is no correlation between Nitric oxide concentrations and proportion of non-retail business acres per town.

Alternate Hypotheis = There is a correlation between Nitric oxide concentrations and proportion of non-retail business acres per town.


```python
scipy.stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])
```

Pearson Coefficient is 0.76365 and p-value less than 0.05, we reject the Null Hypothesis as *'there is a positive correlation between Nitric oxide concentration and proportion of non-retail business acres per town.*'

This was also confirmed in the scatter plot visualised in TASK#4

## TASK#9 What is the impact of an additional weighted distance  to the five Boston employment centres on the median value of owner occupied homes? (Regression analysis)

```python
X = boston_df['DIS']
y = boston_df['MEDV']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

predictions = model.predict()

model.summary()
```

DIS coeff indicates that *'an additional weighted distance to the 5 empolyment centers in Boston causes an increase of 1.0916 the median value of owner occupied homes.*'
